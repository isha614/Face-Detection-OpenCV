```python
import cv2 #OpenCV
import matplotlib.pyplot as plt
```

```python
img = plt.imread('pic.jpg')
```

```python
img.shape
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
model=cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml')
```

```python
faces=model.detectMultiScale(img)
```

```python
faces
```

```python
face=faces[-1]
x,y,w,h=face
```

```python
print(x,y,w,h)
```

```python
img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
plt.imshow(img)
```

```python
for face in faces:
  print(face)
  x,y,w,h=face
  img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(img)
```
